```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/unemployment analysis.csv")
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
df = df.dropna()
df_melted = df.melt(id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Unemployment Rate (%)')
sns.histplot(df_melted['Unemployment Rate (%)'], kde=True,color="red")
plt.title("Distribution of Unemployment Rates (1991-2021)")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.show()

```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Unemployment Rate (%)', data=df_melted)
plt.title("Unemployment Rate Over Time")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
